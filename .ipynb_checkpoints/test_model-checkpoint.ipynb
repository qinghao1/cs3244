```python
# Imports
import keras
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os, glob, random, cv2, sklearn.metrics
import skimage.transform

from model import Model

%matplotlib inline
```

```python
# Load test data
test_images_path = 'data/test_images.npy'
test_label_path = 'data/test_labels.npy'
test_images = np.load(test_images_path)
test_labels = np.load(test_label_path)
print(len(test_images), len(test_labels))
```

```python
# Rescale image to 0..1 as it is now in uint8
test_images *= 1./255
```

```python
# Parameters
IMAGE_DIM = (512,512,)

ROOT_DIR = os.getcwd()
IMAGE_DIR = os.path.join(ROOT_DIR, 'images')
CATEGORIES = [folder for folder in os.listdir(IMAGE_DIR) 
               if os.path.isdir(os.path.join(IMAGE_DIR, folder))]
CATEGORIES.sort() # Alphabetical order
NUM_CLASSES = len(CATEGORIES)
print('{} classes: {}'.format(NUM_CLASSES, CATEGORIES))
```

```python
# Check random images
for _ in range(5):
    random_idx = random.randint(0, len(test_images) - 1)
    rdm_image, rdm_label = test_images[random_idx] * 255, test_labels[random_idx]
    rdm_label = np.argmax(rdm_label, axis=0) # Convert from one-hot to integer
    rdm_image = rdm_image[...,::-1] # Convert from BGR to RGB
    plt.figure()
    plt.title(CATEGORIES[rdm_label])
    plt.imshow(rdm_image)
```

```python
# Load model
MODEL_NAME = "inception-resnet"
model = Model.getCompiledModel(MODEL_NAME, IMAGE_DIM, NUM_CLASSES)
```

```python
# Load trained weights
last_epoch = 272
model.load_weights('{}_weights.{:d}.h5'.format(MODEL_NAME, last_epoch))
```

```python
test_image = cv2.imread("images/hibiscus_rosa_sinensis/5a608220aa3b4b30acd7d1d11cc1312a.jpg")
test_image = skimage.transform.resize(test_image, (512,512))
test_image /= 255
test_image.shape = (1,) + test_image.shape
print(test_image.shape)
test_result = model.predict(test_image)
print(test_result)
test_result = np.argmax(test_result[0], axis=0)
print(test_result)
print(CATEGORIES[test_result])
```

```python
# # Evaluate model with built-in Keras function
# scores = model.evaluate(x=test_images, y=test_labels) #[loss, accuracy]
# print(scores)
```

```python
#Predict
predictions = model.predict(test_images)
```

```python
#Convert results to 1d
predictions1d = np.argmax(predictions, axis=1)
test_labels1d = np.argmax(test_labels, axis=1)
print(predictions1d)
print(test_labels1d)
```

```python
#Confusion matrix
cm = sklearn.metrics.confusion_matrix(test_labels1d, predictions1d)
print(cm)
```

```python
#Normalized confusin matrix where values are in 0..1
normalized_cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
print(normalized_cm)
```

```python
#Make heatmap of CM and normalized CM
fig = plt.figure(figsize = (10,10))
heatmap = sns.heatmap(cm, annot=True, cmap='Blues', cbar=False, annot_kws={'fontsize':14})
heatmap.yaxis.set_ticklabels(CATEGORIES, rotation=0, ha='right')
heatmap.xaxis.set_ticklabels(CATEGORIES, rotation=45, ha='right')
plt.ylabel('True label', fontsize = 14)
plt.xlabel('Predicted label', fontsize = 14)
plt.title('Confusion matrix for {}'.format(MODEL_NAME), fontsize = 16)
plt.savefig('{} confusion matrix'.format(MODEL_NAME))

fig = plt.figure(figsize = (12,10))
heatmap = sns.heatmap(normalized_cm, annot=True, fmt='0.2f', cmap='Blues')
heatmap.yaxis.set_ticklabels(CATEGORIES, rotation=0, ha='right')
heatmap.xaxis.set_ticklabels(CATEGORIES, rotation=45, ha='right')
plt.ylabel('True label', fontsize = 14)
plt.xlabel('Predicted label', fontsize = 14)
plt.title('Normalized confusion matrix for {}'.format(MODEL_NAME), fontsize = 16)
plt.savefig('{} normalized confusion matrix'.format(MODEL_NAME))
```

```python
#Generate and save report
report = sklearn.metrics.classification_report(test_labels1d, predictions1d, target_names=CATEGORIES)
print(report)
f = open('{} classification report.txt'.format(MODEL_NAME), 'w')
f.write(report)
f.close()
```
